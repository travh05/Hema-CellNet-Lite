```python
import glob
import re
import pandas as pd

files = glob.glob("../data/*_log2.csv")
print("Found files:", files)

matrix = None

for path in files:
    df = pd.read_csv(path)
    first_col = df.columns[0]

    gene = re.sub(r"\s*\(log2\).*", "", first_col)

    row = df.iloc[0, 1:]

    row = pd.to_numeric(row, errors="coerce")

    gene_df = pd.DataFrame({gene: row})
    gene_df.index.name = "sample"

    if matrix is None:
        matrix = gene_df
    else:
        matrix = matrix.join(gene_df, how="outer")

matrix["cell_type"] = [s.split(".")[0] for s in matrix.index]

print(matrix.shape)
matrix.head()
```

```python
matrix.to_csv("../data/bloodspot_small_panel.tsv", sep="\t")
```
